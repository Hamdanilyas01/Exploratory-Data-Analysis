```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('retail_store_sales (2).csv')
```

```python
df.info(

)
df.describe()
```

```python
df.head(20)
```

```python
df = df.dropna(subset = ['Price Per Unit', 'Quantity'])
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
df['Discount Applied'].fillna(False,inplace = True)
```

```python
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'], errors='coerce')
```

```python
df.info()
```

```python
df['year'] = df['Transaction Date'].dt.year
df['month'] = df['Transaction Date'].dt.month
```

```python
import seaborn as sns
```

```python
top5 = (
    df.groupby('Category')['Total Spent']
    .sum()
    .nlargest(5)
    .reset_index()
)



top5 = df.groupby('Category')['Total Spent'].sum().nlargest(5).reset_index()


plt.figure(figsize=(10,8))
ax = sns.barplot(data=top5, x='Category', y='Total Spent', hue='Category', palette='pastel', legend=False)


for i, value in enumerate(top5['Total Spent']):
    ax.text(i, value + (0.01*value), f"{value:,.0f}", ha='center', va='bottom')


plt.title("Top 5 Categories by Total Spent", fontsize=14)
plt.xlabel("Category")
plt.ylabel("Total Spent")
sns.despine()
plt.tight_layout()
plt.show()
```

```python

top10 = df.groupby('Item')['Total Spent'].sum().nlargest(10)


colors = sns.color_palette("pastel", len(top10))


plt.figure(figsize=(8,8))
plt.pie(top10, labels=top10.index, autopct='%1.1f%%', startangle=90, colors=colors)


centre_circle = plt.Circle((0,0), 0.70, fc='white')
plt.gca().add_artist(centre_circle)

plt.title("Top 10 Items by Total Spent (Donut Chart)", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
sales_by_year = df.groupby('year')['Total Spent'].sum().reset_index()

# Ensure Year is integer
sales_by_year['year'] = sales_by_year['year'].astype(int)
```

```python



plt.figure(figsize=(8,5))
ax = sns.lineplot(data=sales_by_year, x='year', y='Total Spent', marker='o', color='skyblue')


for x, y in zip(sales_by_year['year'], sales_by_year['Total Spent']):
    ax.text(x, y + (0.01*y), f"{y:,.0f}", ha='center', va='bottom')

plt.title("Total Sales by Year", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Total Spent")


ax.set_xticks(sales_by_year['year'])

sns.despine()
plt.tight_layout()
plt.show()
```

```python

payment_counts = df['Location'].value_counts()


colors = sns.color_palette("pastel", len(payment_counts))


plt.figure(figsize=(6,6))
plt.pie(payment_counts, labels=payment_counts.index, autopct='%1.1f%%', startangle=90, colors=colors)


centre_circle = plt.Circle((0,0), 0.70, fc='white')
plt.gca().add_artist(centre_circle)

plt.title("Transactions by Payment Method")
plt.tight_layout()
plt.show()
```

```python

payment_sales = df.groupby('Payment Method')['Total Spent'].sum().reset_index()

plt.figure(figsize=(8,5))
ax = sns.barplot(data=payment_sales, x='Payment Method', y='Total Spent', palette='pastel')


for i, value in enumerate(payment_sales['Total Spent']):
    ax.text(i, value + (0.01*value), f"{value:,.0f}", ha='center', va='bottom')

plt.title("Total Spent by Payment Method")
plt.xlabel("Payment Method")
plt.ylabel("Total Spent")
sns.despine()
plt.tight_layout()
plt.show()
```

```python
df['Month_Name'] = df['Transaction Date'].dt.month_name()

month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']
df['Month_Name'] = pd.Categorical(df['Month_Name'], categories=month_order, ordered=True)

monthly_category_sales = df.groupby(['Month_Name', 'Category'])['Total Spent'].sum().reset_index()
monthly_category_sales['Month_Name'] = pd.Categorical(monthly_category_sales['Month_Name'], categories=month_order, ordered=True)

plt.figure(figsize=(12,6))

sns.barplot(
    data=monthly_category_sales,
    x='Month_Name',
    y='Total Spent',
    hue='Category',
    palette='viridis'
)

plt.title('Monthly Sales by Category')
plt.xlabel('Month')
plt.ylabel('Total Sales')

plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

df['YearMonth'] = df['Transaction Date'].dt.to_period('M')


sales_by_month = df.groupby('YearMonth')['Total Spent'].sum().reset_index()


sales_by_month['YearMonth'] = sales_by_month['YearMonth'].astype(str)

plt.figure(figsize=(12,6))
ax = sns.lineplot(data=sales_by_month, x='YearMonth', y='Total Spent', marker='o', color='skyblue')


plt.title("Total Sales by Month", fontsize=14)
plt.xlabel("Month")
plt.ylabel("Total Spent")
plt.xticks(rotation=45)
sns.despine()
plt.tight_layout()
plt.show()
```

```python

aov_category = df.groupby('Category')['Total Spent'].mean().sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(y=aov_category.index, x=aov_category.values, palette="viridis")

plt.title("Average Order Value (AOV) by Category", fontsize=14)
plt.xlabel("Average Order Value")
plt.ylabel("Category")
plt.tight_layout()


for i, v in enumerate(aov_category.values):
    plt.text(v, i, f"{v:.2f}", va='center', fontsize=9)

plt.show()
```

```python

top_customers = df.groupby('Customer ID')['Total Spent'].sum().nlargest(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top_customers.index, y=top_customers.values, palette="flare")

plt.title("Top 10 Customers by Total Sales", fontsize=14)
plt.xlabel("Customer ID")
plt.ylabel("Total Spent")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()


for i, v in enumerate(top_customers.values):
    plt.text(i, v, f"{v:.2f}", ha='center', va='bottom', fontsize=9)

plt.show()
```

```python

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix


features = ['Category', 'Item', 'Payment Method', 'Location', 'month', 'year']
target = 'Discount Applied'


X = df[features]
y = df[target]


if y.dtype == 'object' or y.dtype.name == 'bool':
    le = LabelEncoder()
    y = le.fit_transform(y)


X_encoded = pd.get_dummies(X, drop_first=True)


X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42, stratify=y
)


rf_model = RandomForestClassifier(n_estimators=10, random_state=42)
cv_scores = cross_val_score(rf_model, X_train, y_train, cv=5, scoring='accuracy')

print("Cross-validation scores:", cv_scores)
print("Mean CV accuracy:", cv_scores.mean())


rf_model.fit(X_train, y_train)


y_pred = rf_model.predict(X_test)

print("\nTest Set Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))


cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=le.classes_, yticklabels=le.classes_)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - Discount Applied')
plt.show()


importances = rf_model.feature_importances_
feature_names = X_encoded.columns

feat_imp = pd.Series(importances, index=feature_names).sort_values(ascending=False).head(5)

plt.figure(figsize=(10,6))
sns.barplot(x=feat_imp.values, y=feat_imp.index, palette='viridis')
plt.title("Top 5 Features Influencing Discount Applied")
plt.xlabel("Feature Importance")
plt.ylabel("Feature")
plt.tight_layout()
plt.show()
```
